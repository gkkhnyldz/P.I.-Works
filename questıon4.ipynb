```python

import numpy as np
import pandas as pd
ourdata = pd.read_csv("country_vaccination_stats.csv")
def impute_by_min(country):
    country__data = ourdata[ourdata["country"] == country]
    min_vaccinations = country__data["daily_vaccinations"].min()
    ourdata.loc[(ourdata["country"] == country) & (ourdata["daily_vaccinations"].isna()), "daily_vaccinations"] = min_vaccinations
    ourdata.loc[(ourdata["country"] == country) & (ourdata["daily_vaccinations"].isnull()), "daily_vaccinations"] = 0

countries = ourdata["country"].unique()
for country in countries:
    impute_by_min(country)
```
